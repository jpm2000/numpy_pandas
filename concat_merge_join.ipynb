```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.DataFrame({
    "A": ['A0', 'A1', 'A2', 'A3'],
    'B': ['B0', 'B1', 'B2', 'B3'],
    'C': ['C0', 'C1', 'C2', 'C3'],
    'D': ['D0', 'D1', 'D2', 'D3']
})

df2 = pd.DataFrame({
    "A": ['A4', 'A5', 'A6', 'A7'],
    'B': ['B4', 'B5', 'B6', 'B7'],
    'C': ['C4', 'C5', 'C6', 'C7'],
    'D': ['D4', 'D5', 'D6', 'D7']
})
```

```python
# El problema con los indices, heredan los indices anteriores
pd.concat([df1, df2], ignore_index=True)
```

```python
# Por columnas, nos es muy común hacerlo de esta manera
pd.concat([df1, df2], axis=1)
```

Merge

```python
df3 = pd.DataFrame({
    'key': ['k0', 'k1', 'k2', 'k3'],
    "A": ['A0', 'A1', 'A2', 'A3'],
    'B': ['B0', 'B1', 'B2', 'B3']
})

df4 = pd.DataFrame(
    {
        'key': ['k0', 'k1', 'k2', 'k3'],
        'C': ['C0', 'C1', 'C2', 'C3'],
        'D': ['D0', 'D1', 'D2', 'D3']
    }
)
```

```python
df3.merge(df4, on='key')
```

```python
izq = pd.DataFrame({
    'key': ['k0', 'k1', 'k2', 'k3'],
    "A": ['A0', 'A1', 'A2', 'A3'],
    'B': ['B0', 'B1', 'B2', 'B3']
})

der = pd.DataFrame(
    {
        'key2': ['k0', 'k1', 'k2', 'k3'],
        'C': ['C0', 'C1', 'C2', 'C3'],
        'D': ['D0', 'D1', 'D2', 'D3']
    }
)
```

```python
# Sabe que pueden haber diferencias por las keys, pero igual hace la relación
izq.merge(der, left_on='key', right_on='key2')
```

```python
izq = pd.DataFrame({
    'key': ['k0', 'k1', 'k2', 'k3'],
    "A": ['A0', 'A1', 'A2', 'A3'],
    'B': ['B0', 'B1', 'B2', 'B3']
})

der = pd.DataFrame(
    {
        'key2': ['k0', 'k1', 'k2', np.nan],
        'C': ['C0', 'C1', 'C2', 'C3'],
        'D': ['D0', 'D1', 'D2', 'D3']
    }
)
```

```python
izq
```

```python
der
```

```python
# No me traer el tercer key, porque no hace match. Inner join, solo trae las coincidenccias
izq.merge(der, left_on='key', right_on='key2')
```

```python
# Si quiero traer valores que sean o no coincidencia 
izq.merge(der, left_on='key', right_on='key2', how='left')
```

```python
# Si quiero traer valores que sean o no coincidencia 
izq.merge(der, left_on='key', right_on='key2', how='right')
```

```python
# Si quiero traer valores que sean o no coincidencia 
izq.merge(der, left_on='key', right_on='key2', how='inner')
```
