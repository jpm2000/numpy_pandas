```python
books = pd.read_csv('/Users/juanmanrique/Documents/numpy_pandas/bestsellers with categories.csv', sep=',', header=0)
```

```python
# Pivot
books.pivot_table(index='Author', columns='Genre', values='User Rating')
```

```python
books.pivot_table(index='Genre', columns='Year', values='User Rating', aggfunc='sum')
```

```python
books.pivot_table(index='Genre', columns='Year', values='User Rating', aggfunc='median')
```

```python
books.pivot_table(index='Genre', columns='Year', values='User Rating', aggfunc='count')
```
